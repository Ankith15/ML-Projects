# Preprocessing the data

## Scaling the data

```python
def process(image,label):
    image = tf.cast(image/255. ,tf.float32)
    return image,label

train_ds = train_ds.map(process)
```

# Splitting the data

```python
len(train_ds)
```

```python
train_size = int(len(train_ds)*.7)
val_size = int(len(train_ds)*.2)+1
test_size = int(len(train_ds)*.1)
```

```python
train = train_ds.take(train_size)
val = train_ds.skip(train_size).take(val_size)
test = train_ds.skip(train_size+val_size).take(test_size)
```

# Model Building

```python
from keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dense,Flatten,Dropout
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout
```

```python
model=Sequential()
```

```python
# create CNN model

model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(150,150,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(8,activation='softmax'))
```

```python
model.compile('adam', loss=tf.losses.CategoricalCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist=model.fit(train,epochs=10,validation_data=val)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten
from keras.applications.vgg16 import VGG16
```

```python
conv_base = VGG16(
    weights='imagenet',
    include_top = False,
    input_shape=(150,150,3)
)
```

```python
conv_base.summary()
```

```python
model = Sequential()

model.add(conv_base)
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
conv_base.trainable = False
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(train,epochs=10,validation_data=val)
```
